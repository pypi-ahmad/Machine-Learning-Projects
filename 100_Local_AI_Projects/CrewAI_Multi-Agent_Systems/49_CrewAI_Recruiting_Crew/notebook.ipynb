## Resume screening → interview design → evaluation → recommendation

**Agents:** Resume Screener → Interview Designer → Technical Evaluator → Hiring Advisor

**Stack:** CrewAI · Ollama · Jupyter

## Step 1 — Setup Local LLM

```python
from langchain_community.llms import Ollama

llm = Ollama(model="qwen3:8b", temperature=0.3)
print("Local Ollama LLM ready for CrewAI!")
```

## Step 2 — Define Agents

```python
from crewai import Agent

resume_screener = Agent(
    role='Resume Screener',
    goal='Screen the resume against job requirements',
    backstory='Screen the resume against job requirements. Score on technical skills, experience level, and cultural indicators.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

interview_designer = Agent(
    role='Interview Designer',
    goal='Design a structured interview with specific questions targeting identified skill areas and gaps',
    backstory='Design a structured interview with specific questions targeting identified skill areas and gaps.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

technical_evaluator = Agent(
    role='Technical Evaluator',
    goal='Create a take-home technical assessment aligned with the role',
    backstory='Create a take-home technical assessment aligned with the role. Include rubric and evaluation criteria.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

hiring_advisor = Agent(
    role='Hiring Advisor',
    goal='Provide a hiring recommendation based on all inputs',
    backstory='Provide a hiring recommendation based on all inputs. Include risk factors and onboarding suggestions.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

print("Agents defined: Resume Screener, Interview Designer, Technical Evaluator, Hiring Advisor")
```

## Step 3 — Define Tasks

```python
from crewai import Task

resume_screening_task = Task(
    description='Screen the resume against job requirements. Score on technical skills, experience level, and cultural indicators.',
    expected_output='Detailed resume screening',
    agent=resume_screener,
)

interview_plan_task = Task(
    description='Design a structured interview with specific questions targeting identified skill areas and gaps.',
    expected_output='Detailed interview plan',
    agent=interview_designer,
)

technical_assessment_task = Task(
    description='Create a take-home technical assessment aligned with the role. Include rubric and evaluation criteria.',
    expected_output='Detailed technical assessment',
    agent=technical_evaluator,
)

hiring_recommendation_task = Task(
    description='Provide a hiring recommendation based on all inputs. Include risk factors and onboarding suggestions.',
    expected_output='Detailed hiring recommendation',
    agent=hiring_advisor,
)

tasks = [resume_screening_task, interview_plan_task, technical_assessment_task, hiring_recommendation_task]
print(f"{len(tasks)} tasks defined")
```

## Step 4 — Assemble and Run Crew

```python
from crewai import Crew, Process

crew = Crew(
    agents=[resume_screener, interview_designer, technical_evaluator, hiring_advisor],
    tasks=tasks,
    process=Process.sequential,
    verbose=True,
)

print("Crew assembled! Running CrewAI Recruiting Crew...")
result = crew.kickoff(inputs={"input": "Evaluate a candidate for Senior Backend Engineer: 5 years Python, Django, PostgreSQL, some Kubernetes, previously at a startup."})
```

## Step 5 — Review Results

```python
print("="*60)
print("CREW OUTPUT")
print("="*60)
print(result)
```
