## Step 1 — Setup Local LLM

```python
from langchain_community.llms import Ollama

llm = Ollama(model="qwen3:8b", temperature=0.3)
print("Local Ollama LLM ready for CrewAI!")
```

## Step 2 — Define Agents

```python
from crewai import Agent

product_manager = Agent(
    role='Product Manager',
    goal='Create a product requirements document (PRD) for the launch including target users, key features, success metrics, and timeline',
    backstory='Create a product requirements document (PRD) for the launch including target users, key features, success metrics, and timeline.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

marketing_strategist = Agent(
    role='Marketing Strategist',
    goal='Develop a go-to-market strategy with messaging, channels, content calendar, and launch day plan',
    backstory='Develop a go-to-market strategy with messaging, channels, content calendar, and launch day plan.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

data_analyst = Agent(
    role='Data Analyst',
    goal='Define the analytics plan: key metrics, dashboards needed, A/B tests to run, and tracking requirements',
    backstory='Define the analytics plan: key metrics, dashboards needed, A/B tests to run, and tracking requirements.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

qa_lead = Agent(
    role='QA Lead',
    goal='Create a launch readiness checklist: testing scenarios, rollback plan, monitoring alerts, and support escalation process',
    backstory='Create a launch readiness checklist: testing scenarios, rollback plan, monitoring alerts, and support escalation process.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

print("Agents defined: Product Manager, Marketing Strategist, Data Analyst, QA Lead")
```

## Step 3 — Define Tasks

```python
from crewai import Task

prd_task = Task(
    description='Create a product requirements document (PRD) for the launch including target users, key features, success metrics, and timeline.',
    expected_output='Detailed prd',
    agent=product_manager,
)

gtm_strategy_task = Task(
    description='Develop a go-to-market strategy with messaging, channels, content calendar, and launch day plan.',
    expected_output='Detailed gtm strategy',
    agent=marketing_strategist,
)

analytics_plan_task = Task(
    description='Define the analytics plan: key metrics, dashboards needed, A/B tests to run, and tracking requirements.',
    expected_output='Detailed analytics plan',
    agent=data_analyst,
)

launch_checklist_task = Task(
    description='Create a launch readiness checklist: testing scenarios, rollback plan, monitoring alerts, and support escalation process.',
    expected_output='Detailed launch checklist',
    agent=qa_lead,
)

tasks = [prd_task, gtm_strategy_task, analytics_plan_task, launch_checklist_task]
print(f"{len(tasks)} tasks defined")
```

## Step 4 — Assemble and Run Crew

```python
from crewai import Crew, Process

crew = Crew(
    agents=[product_manager, marketing_strategist, data_analyst, qa_lead],
    tasks=tasks,
    process=Process.sequential,
    verbose=True,
)

print("Crew assembled! Running CrewAI Product Launch Crew...")
result = crew.kickoff(inputs={"input": "Launch a new AI writing assistant feature for an existing productivity SaaS platform."})
```

## Step 5 — Review Results
