## Validate a startup idea with multi-agent analysis

**Agents:** Market Researcher → Competitor Analyst → Financial Analyst → Devil's Advocate

**Stack:** CrewAI · Ollama · Jupyter

## Step 1 — Setup Local LLM

```python
from langchain_community.llms import Ollama

llm = Ollama(model="qwen3:8b", temperature=0.3)
print("Local Ollama LLM ready for CrewAI!")
```

## Step 2 — Define Agents

```python
from crewai import Agent

market_researcher = Agent(
    role='Market Researcher',
    goal='Research the market size, trends, and target audience for the startup idea',
    backstory='Research the market size, trends, and target audience for the startup idea. Provide data-driven insights.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

competitor_analyst = Agent(
    role='Competitor Analyst',
    goal='Identify top 5 competitors, their strengths/weaknesses, pricing, and market positioning',
    backstory='Identify top 5 competitors, their strengths/weaknesses, pricing, and market positioning.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

financial_analyst = Agent(
    role='Financial Analyst',
    goal='Estimate startup costs, revenue projections, unit economics, and break-even timeline',
    backstory='Estimate startup costs, revenue projections, unit economics, and break-even timeline.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

devils_advocate = Agent(
    role='Devil's Advocate',
    goal='Challenge the startup idea with tough questions, risks, and potential failure modes',
    backstory='Challenge the startup idea with tough questions, risks, and potential failure modes.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

print("Agents defined: Market Researcher, Competitor Analyst, Financial Analyst, Devil's Advocate")
```

## Step 3 — Define Tasks

```python
from crewai import Task

market_analysis_task = Task(
    description='Research the market size, trends, and target audience for the startup idea. Provide data-driven insights.',
    expected_output='Detailed market analysis',
    agent=market_researcher,
)

competitor_report_task = Task(
    description='Identify top 5 competitors, their strengths/weaknesses, pricing, and market positioning.',
    expected_output='Detailed competitor report',
    agent=competitor_analyst,
)

financial_model_task = Task(
    description='Estimate startup costs, revenue projections, unit economics, and break-even timeline.',
    expected_output='Detailed financial model',
    agent=financial_analyst,
)

risk_assessment_task = Task(
    description='Challenge the startup idea with tough questions, risks, and potential failure modes.',
    expected_output='Detailed risk assessment',
    agent=devils_advocate,
)

tasks = [market_analysis_task, competitor_report_task, financial_model_task, risk_assessment_task]
print(f"{len(tasks)} tasks defined")
```

## Step 4 — Assemble and Run Crew

```python
from crewai import Crew, Process

crew = Crew(
    agents=[market_researcher, competitor_analyst, financial_analyst, devils_advocate],
    tasks=tasks,
    process=Process.sequential,
    verbose=True,
)

print("Crew assembled! Running CrewAI Startup Validation Crew...")
result = crew.kickoff(inputs={"input": "Validate this startup idea: An AI-powered local tutoring platform that matches students with tutors based on learning style."})
```

## Step 5 — Review Results

```python
print("="*60)
print("CREW OUTPUT")
print("="*60)
print(result)
```
