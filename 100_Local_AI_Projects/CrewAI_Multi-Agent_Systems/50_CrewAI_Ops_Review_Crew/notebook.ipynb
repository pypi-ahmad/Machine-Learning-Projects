## Step 1 — Setup Local LLM

```python
from langchain_community.llms import Ollama

llm = Ollama(model="qwen3:8b", temperature=0.3)
print("Local Ollama LLM ready for CrewAI!")
```

## Step 2 — Define Agents

```python
from crewai import Agent

ops_analyst = Agent(
    role='Ops Analyst',
    goal='Analyze the operational data and identify key trends: uptime, latency, error rates, resource utilization',
    backstory='Analyze the operational data and identify key trends: uptime, latency, error rates, resource utilization.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

risk_reviewer = Agent(
    role='Risk Reviewer',
    goal='Identify operational risks from the analysis',
    backstory='Identify operational risks from the analysis. Classify by probability and impact. Recommend mitigations.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

optimization_advisor = Agent(
    role='Optimization Advisor',
    goal='Suggest specific optimizations for infrastructure, processes, and tooling based on the analysis',
    backstory='Suggest specific optimizations for infrastructure, processes, and tooling based on the analysis.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

report_writer = Agent(
    role='Report Writer',
    goal='Write an executive operations review report summarizing findings, risks, and recommended actions',
    backstory='Write an executive operations review report summarizing findings, risks, and recommended actions.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

print("Agents defined: Ops Analyst, Risk Reviewer, Optimization Advisor, Report Writer")
```

## Step 3 — Define Tasks

```python
from crewai import Task

ops_analysis_task = Task(
    description='Analyze the operational data and identify key trends: uptime, latency, error rates, resource utilization.',
    expected_output='Detailed ops analysis',
    agent=ops_analyst,
)

risk_review_task = Task(
    description='Identify operational risks from the analysis. Classify by probability and impact. Recommend mitigations.',
    expected_output='Detailed risk review',
    agent=risk_reviewer,
)

optimization_plan_task = Task(
    description='Suggest specific optimizations for infrastructure, processes, and tooling based on the analysis.',
    expected_output='Detailed optimization plan',
    agent=optimization_advisor,
)

ops_report_task = Task(
    description='Write an executive operations review report summarizing findings, risks, and recommended actions.',
    expected_output='Detailed ops report',
    agent=report_writer,
)

tasks = [ops_analysis_task, risk_review_task, optimization_plan_task, ops_report_task]
print(f"{len(tasks)} tasks defined")
```

## Step 4 — Assemble and Run Crew

```python
from crewai import Crew, Process

crew = Crew(
    agents=[ops_analyst, risk_reviewer, optimization_advisor, report_writer],
    tasks=tasks,
    process=Process.sequential,
    verbose=True,
)

print("Crew assembled! Running CrewAI Ops Review Crew...")
result = crew.kickoff(inputs={"input": "Review Q4 operations: 99.8% uptime, p95 latency 450ms (target 200ms), 3 incidents, infrastructure costs up 25%."})
```
