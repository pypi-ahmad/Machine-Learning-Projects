## Step 1 — Setup Local LLM

```python
from langchain_community.llms import Ollama

llm = Ollama(model="qwen3:8b", temperature=0.3)
print("Local Ollama LLM ready for CrewAI!")
```

## Step 2 — Define Agents

```python
from crewai import Agent

jd_analyzer = Agent(
    role='JD Analyzer',
    goal='Parse the job description',
    backstory='Parse the job description. Extract must-have skills, nice-to-haves, company values, and culture signals.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

resume_tailor = Agent(
    role='Resume Tailor',
    goal='Rewrite the resume to emphasize relevant experience matching the JD requirements',
    backstory='Rewrite the resume to emphasize relevant experience matching the JD requirements. Use action verbs and metrics.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

interview_coach = Agent(
    role='Interview Coach',
    goal='Generate 10 likely interview questions based on the JD and company',
    backstory='Generate 10 likely interview questions based on the JD and company. Include behavioral, technical, and situational questions.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

strategy_advisor = Agent(
    role='Strategy Advisor',
    goal='Provide a comprehensive job application strategy: networking tips, portfolio suggestions, and follow-up plan',
    backstory='Provide a comprehensive job application strategy: networking tips, portfolio suggestions, and follow-up plan.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

print("Agents defined: JD Analyzer, Resume Tailor, Interview Coach, Strategy Advisor")
```

## Step 3 — Define Tasks

```python
from crewai import Task

jd_analysis_task = Task(
    description='Parse the job description. Extract must-have skills, nice-to-haves, company values, and culture signals.',
    expected_output='Detailed jd analysis',
    agent=jd_analyzer,
)

tailored_resume_task = Task(
    description='Rewrite the resume to emphasize relevant experience matching the JD requirements. Use action verbs and metrics.',
    expected_output='Detailed tailored resume',
    agent=resume_tailor,
)

interview_prep_task = Task(
    description='Generate 10 likely interview questions based on the JD and company. Include behavioral, technical, and situational questions.',
    expected_output='Detailed interview prep',
    agent=interview_coach,
)

application_strategy_task = Task(
    description='Provide a comprehensive job application strategy: networking tips, portfolio suggestions, and follow-up plan.',
    expected_output='Detailed application strategy',
    agent=strategy_advisor,
)

tasks = [jd_analysis_task, tailored_resume_task, interview_prep_task, application_strategy_task]
print(f"{len(tasks)} tasks defined")
```

## Step 4 — Assemble and Run Crew

```python
from crewai import Crew, Process

crew = Crew(
    agents=[jd_analyzer, resume_tailor, interview_coach, strategy_advisor],
    tasks=tasks,
    process=Process.sequential,
    verbose=True,
)

print("Crew assembled! Running CrewAI Job Hunt Crew...")
result = crew.kickoff(inputs={"input": "Prepare for a Senior ML Engineer role at a fintech startup. Resume: 4 years Python, ML models, deployed production systems."})
```
