## Step 1 — Setup Local LLM

```python
from langchain_community.llms import Ollama

llm = Ollama(model="qwen3:8b", temperature=0.3)
print("Local Ollama LLM ready for CrewAI!")
```

## Step 2 — Define Agents

```python
from crewai import Agent

icp_analyst = Agent(
    role='ICP Analyst',
    goal='Define the Ideal Customer Profile for an AI analytics SaaS product',
    backstory='Define the Ideal Customer Profile for an AI analytics SaaS product. Include industry, company size, tech stack, pain points.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

company_researcher = Agent(
    role='Company Researcher',
    goal='Research 3 target companies matching the ICP',
    backstory='Research 3 target companies matching the ICP. Find key decision makers, recent initiatives, and potential hooks.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

personalization_specialist = Agent(
    role='Personalization Specialist',
    goal='Create personalized talking points for each company, connecting our product to their specific needs',
    backstory='Create personalized talking points for each company, connecting our product to their specific needs.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

email_copywriter = Agent(
    role='Email Copywriter',
    goal='Draft a personalized cold outreach email for each company',
    backstory='Draft a personalized cold outreach email for each company. Under 150 words. Clear CTA for a 15-min demo.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

print("Agents defined: ICP Analyst, Company Researcher, Personalization Specialist, Email Copywriter")
```

## Step 3 — Define Tasks

```python
from crewai import Task

icp_definition_task = Task(
    description='Define the Ideal Customer Profile for an AI analytics SaaS product. Include industry, company size, tech stack, pain points.',
    expected_output='Detailed icp definition',
    agent=icp_analyst,
)

company_profiles_task = Task(
    description='Research 3 target companies matching the ICP. Find key decision makers, recent initiatives, and potential hooks.',
    expected_output='Detailed company profiles',
    agent=company_researcher,
)

personalization_notes_task = Task(
    description='Create personalized talking points for each company, connecting our product to their specific needs.',
    expected_output='Detailed personalization notes',
    agent=personalization_specialist,
)

outreach_emails_task = Task(
    description='Draft a personalized cold outreach email for each company. Under 150 words. Clear CTA for a 15-min demo.',
    expected_output='Detailed outreach emails',
    agent=email_copywriter,
)

tasks = [icp_definition_task, company_profiles_task, personalization_notes_task, outreach_emails_task]
print(f"{len(tasks)} tasks defined")
```

## Step 4 — Assemble and Run Crew

```python
from crewai import Crew, Process

crew = Crew(
    agents=[icp_analyst, company_researcher, personalization_specialist, email_copywriter],
    tasks=tasks,
    process=Process.sequential,
    verbose=True,
)

print("Crew assembled! Running CrewAI Lead Gen Crew...")
result = crew.kickoff(inputs={"input": "Generate outreach for an AI-powered analytics platform targeting mid-market SaaS companies."})
```
