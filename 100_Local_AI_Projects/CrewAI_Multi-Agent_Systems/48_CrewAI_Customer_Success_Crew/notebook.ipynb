**Stack:** CrewAI · Ollama · Jupyter

## Step 1 — Setup Local LLM

```python
from langchain_community.llms import Ollama

llm = Ollama(model="qwen3:8b", temperature=0.3)
print("Local Ollama LLM ready for CrewAI!")
```

## Step 2 — Define Agents

```python
from crewai import Agent

complaint_analyst = Agent(
    role='Complaint Analyst',
    goal='Analyze the customer complaint',
    backstory="Analyze the customer complaint. Classify severity, identify root cause, and determine if it is a systemic issue.",
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

churn_risk_assessor = Agent(
    role='Churn Risk Assessor',
    goal='Evaluate churn risk based on the complaint, account history, and revenue importance',
    backstory='Evaluate churn risk based on the complaint, account history, and revenue importance. Score 1-10.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

response_writer = Agent(
    role='Response Writer',
    goal='Draft a personalized response to the customer',
    backstory='Draft a personalized response to the customer. Acknowledge the issue, provide timeline, offer compensation if appropriate.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

action_planner = Agent(
    role='Action Planner',
    goal='Create an internal action plan to resolve the issue and prevent recurrence',
    backstory='Create an internal action plan to resolve the issue and prevent recurrence. Include owners and deadlines.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

print("Agents defined: Complaint Analyst, Churn Risk Assessor, Response Writer, Action Planner")
```

## Step 3 — Define Tasks

```python
from crewai import Task

complaint_analysis_task = Task(
    description="Analyze the customer complaint. Classify severity, identify root cause, and determine if it is a systemic issue.",
    expected_output='Detailed complaint analysis',
    agent=complaint_analyst,
)

churn_assessment_task = Task(
    description='Evaluate churn risk based on the complaint, account history, and revenue importance. Score 1-10.',
    expected_output='Detailed churn assessment',
    agent=churn_risk_assessor,
)

customer_response_task = Task(
    description='Draft a personalized response to the customer. Acknowledge the issue, provide timeline, offer compensation if appropriate.',
    expected_output='Detailed customer response',
    agent=response_writer,
)

action_plan_task = Task(
    description='Create an internal action plan to resolve the issue and prevent recurrence. Include owners and deadlines.',
    expected_output='Detailed action plan',
    agent=action_planner,
)

tasks = [complaint_analysis_task, churn_assessment_task, customer_response_task, action_plan_task]
print(f"{len(tasks)} tasks defined")
```

## Step 4 — Assemble and Run Crew

```python
from crewai import Crew, Process

crew = Crew(
    agents=[complaint_analyst, churn_risk_assessor, response_writer, action_planner],
    tasks=tasks,
    process=Process.sequential,
    verbose=True,
)

print("Crew assembled! Running CrewAI Customer Success Crew...")
result = crew.kickoff(inputs={"input": "Customer complaint: Your API has been down 3 times this month. We are paying $5000/month and considering switching to a competitor."})
```
