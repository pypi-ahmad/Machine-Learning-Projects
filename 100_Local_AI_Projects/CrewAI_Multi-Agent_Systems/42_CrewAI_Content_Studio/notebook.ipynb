# Project 42 — CrewAI Content Studio
## Multi-agent content creation pipeline: research → write → edit → repurpose

**Agents:** Content Researcher → Content Writer → Editor → Content Repurposer

**Stack:** CrewAI · Ollama · Jupyter

```python
# !pip install -q crewai langchain-community
```

## Step 1 — Setup Local LLM

```python
from langchain_community.llms import Ollama

llm = Ollama(model="qwen3:8b", temperature=0.3)
print("Local Ollama LLM ready for CrewAI!")
```

## Step 2 — Define Agents

```python
from crewai import Agent

content_researcher = Agent(
    role='Content Researcher',
    goal='Research the topic thoroughly',
    backstory='Research the topic thoroughly. Find key facts, statistics, expert opinions, and trending angles.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

content_writer = Agent(
    role='Content Writer',
    goal='Write a compelling 800-word blog post based on the research brief',
    backstory='Write a compelling 800-word blog post based on the research brief. Use clear structure with headers.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

editor = Agent(
    role='Editor',
    goal='Edit the draft for clarity, grammar, tone consistency, and engagement',
    backstory='Edit the draft for clarity, grammar, tone consistency, and engagement. Improve the hook and CTA.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

content_repurposer = Agent(
    role='Content Repurposer',
    goal='Create 3 derivative pieces: a Twitter thread (5 tweets), LinkedIn post, and email newsletter summary',
    backstory='Create 3 derivative pieces: a Twitter thread (5 tweets), LinkedIn post, and email newsletter summary.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

print("Agents defined: Content Researcher, Content Writer, Editor, Content Repurposer")
```

## Step 3 — Define Tasks

```python
from crewai import Task

research_brief_task = Task(
    description='Research the topic thoroughly. Find key facts, statistics, expert opinions, and trending angles.',
    expected_output='Detailed research brief',
    agent=content_researcher,
)

blog_draft_task = Task(
    description='Write a compelling 800-word blog post based on the research brief. Use clear structure with headers.',
    expected_output='Detailed blog draft',
    agent=content_writer,
)

edited_article_task = Task(
    description='Edit the draft for clarity, grammar, tone consistency, and engagement. Improve the hook and CTA.',
    expected_output='Detailed edited article',
    agent=editor,
)

content_variants_task = Task(
    description='Create 3 derivative pieces: a Twitter thread (5 tweets), LinkedIn post, and email newsletter summary.',
    expected_output='Detailed content variants',
    agent=content_repurposer,
)

tasks = [research_brief_task, blog_draft_task, edited_article_task, content_variants_task]
print(f"{len(tasks)} tasks defined")
```

## Step 4 — Assemble and Run Crew

```python
from crewai import Crew, Process

crew = Crew(
    agents=[content_researcher, content_writer, editor, content_repurposer],
    tasks=tasks,
    process=Process.sequential,
    verbose=True,
)

print("Crew assembled! Running CrewAI Content Studio...")
result = crew.kickoff(inputs={"input": "Topic: How small businesses can use AI to compete with enterprises in 2025"})
```

## Step 5 — Review Results
