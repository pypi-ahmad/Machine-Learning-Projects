## Step 1 — Setup Local LLM

```python
from langchain_community.llms import Ollama

llm = Ollama(model="qwen3:8b", temperature=0.3)
print("Local Ollama LLM ready for CrewAI!")
```

## Step 2 — Define Agents

```python
from crewai import Agent

literature_scout = Agent(
    role='Literature Scout',
    goal='Find and summarize key research papers on the topic',
    backstory='Find and summarize key research papers on the topic. Include authors, year, methodology, and findings.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

methodology_analyst = Agent(
    role='Methodology Analyst',
    goal='Compare research methodologies across papers',
    backstory='Compare research methodologies across papers. Identify common approaches, datasets, and evaluation metrics.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

gap_finder = Agent(
    role='Gap Finder',
    goal='Identify gaps in the existing research',
    backstory='Identify gaps in the existing research. What questions remain unanswered? Where are the opportunities?',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

bibliography_curator = Agent(
    role='Bibliography Curator',
    goal='Create an organized bibliography with APA citations and annotated summaries for each source',
    backstory='Create an organized bibliography with APA citations and annotated summaries for each source.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

print("Agents defined: Literature Scout, Methodology Analyst, Gap Finder, Bibliography Curator")
```

## Step 3 — Define Tasks

```python
from crewai import Task

literature_review_task = Task(
    description='Find and summarize key research papers on the topic. Include authors, year, methodology, and findings.',
    expected_output='Detailed literature review',
    agent=literature_scout,
)

methodology_analysis_task = Task(
    description='Compare research methodologies across papers. Identify common approaches, datasets, and evaluation metrics.',
    expected_output='Detailed methodology analysis',
    agent=methodology_analyst,
)

research_gaps_task = Task(
    description='Identify gaps in the existing research. What questions remain unanswered? Where are the opportunities?',
    expected_output='Detailed research gaps',
    agent=gap_finder,
)

annotated_bibliography_task = Task(
    description='Create an organized bibliography with APA citations and annotated summaries for each source.',
    expected_output='Detailed annotated bibliography',
    agent=bibliography_curator,
)

tasks = [literature_review_task, methodology_analysis_task, research_gaps_task, annotated_bibliography_task]
print(f"{len(tasks)} tasks defined")
```

## Step 4 — Assemble and Run Crew

```python
from crewai import Crew, Process

crew = Crew(
    agents=[literature_scout, methodology_analyst, gap_finder, bibliography_curator],
    tasks=tasks,
    process=Process.sequential,
    verbose=True,
)

print("Crew assembled! Running CrewAI Academic Research Crew...")
result = crew.kickoff(inputs={"input": "Research topic: Effectiveness of RAG systems versus fine-tuning for domain-specific NLP tasks"})
```
