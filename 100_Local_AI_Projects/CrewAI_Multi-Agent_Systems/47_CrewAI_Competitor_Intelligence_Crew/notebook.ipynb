## Step 1 — Setup Local LLM

```python
from langchain_community.llms import Ollama

llm = Ollama(model="qwen3:8b", temperature=0.3)
print("Local Ollama LLM ready for CrewAI!")
```

## Step 2 — Define Agents

```python
from crewai import Agent

feature_analyst = Agent(
    role='Feature Analyst',
    goal='Compare features across the top 3 competitors',
    backstory='Compare features across the top 3 competitors. Create a feature matrix showing strengths and gaps.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

pricing_analyst = Agent(
    role='Pricing Analyst',
    goal='Analyze competitor pricing models',
    backstory='Analyze competitor pricing models. Compare tiers, value per dollar, and positioning strategy.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

swot_analyst = Agent(
    role='SWOT Analyst',
    goal='Conduct SWOT analysis for each competitor',
    backstory='Conduct SWOT analysis for each competitor. Identify our relative advantages and vulnerabilities.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

strategy_summarizer = Agent(
    role='Strategy Summarizer',
    goal='Write an executive memo summarizing competitive landscape and recommending strategic actions',
    backstory='Write an executive memo summarizing competitive landscape and recommending strategic actions.',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

print("Agents defined: Feature Analyst, Pricing Analyst, SWOT Analyst, Strategy Summarizer")
```

## Step 3 — Define Tasks

```python
from crewai import Task

feature_matrix_task = Task(
    description='Compare features across the top 3 competitors. Create a feature matrix showing strengths and gaps.',
    expected_output='Detailed feature matrix',
    agent=feature_analyst,
)

pricing_analysis_task = Task(
    description='Analyze competitor pricing models. Compare tiers, value per dollar, and positioning strategy.',
    expected_output='Detailed pricing analysis',
    agent=pricing_analyst,
)

swot_analysis_task = Task(
    description='Conduct SWOT analysis for each competitor. Identify our relative advantages and vulnerabilities.',
    expected_output='Detailed swot analysis',
    agent=swot_analyst,
)

strategic_memo_task = Task(
    description='Write an executive memo summarizing competitive landscape and recommending strategic actions.',
    expected_output='Detailed strategic memo',
    agent=strategy_summarizer,
)

tasks = [feature_matrix_task, pricing_analysis_task, swot_analysis_task, strategic_memo_task]
print(f"{len(tasks)} tasks defined")
```

## Step 4 — Assemble and Run Crew

```python
from crewai import Crew, Process

crew = Crew(
    agents=[feature_analyst, pricing_analyst, swot_analyst, strategy_summarizer],
    tasks=tasks,
    process=Process.sequential,
    verbose=True,
)

print("Crew assembled! Running CrewAI Competitor Intelligence Crew...")
result = crew.kickoff(inputs={"input": "Analyze competitors for a local-first AI code assistant tool competing with GitHub Copilot, Cursor, and Codeium."})
```
