**Workflow:** Define Requirements → Gather Evidence → Assess Gaps → Generate Report

## Step 1 — Setup

```python
from langgraph.graph import StateGraph, END
from typing import TypedDict
from langchain_ollama import ChatOllama
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm = ChatOllama(model="qwen3:8b", temperature=0.2)

class WorkflowState(TypedDict):
    input_data: str
    accumulated: str
    final_output: str
```

## Step 2 — Define Workflow Nodes

```python
def define_requirements(state: WorkflowState) -> WorkflowState:
    prompt = ChatPromptTemplate.from_messages([
        ("system", "You are step 1 of a LangGraph Compliance Checklist Flow pipeline. "
         "Your role: define requirements. "
         "Process the input and produce structured output for the next step."),
        ("human", "Input: {input}\nPrevious steps: {context}")
    ])
    chain = prompt | llm | StrOutputParser()
    result = chain.invoke({"input": state["input_data"], "context": state.get("accumulated", "")})
    accumulated = state.get("accumulated", "") + f"\n\n[define_requirements]: " + result
    print(f"  Step 1/4: define_requirements")
    return {"accumulated": accumulated, "final_output": result}

def gather_evidence(state: WorkflowState) -> WorkflowState:
    prompt = ChatPromptTemplate.from_messages([
        ("system", "You are step 2 of a LangGraph Compliance Checklist Flow pipeline. "
         "Your role: gather evidence. "
         "Process the input and produce structured output for the next step."),
        ("human", "Input: {input}\nPrevious steps: {context}")
    ])
    chain = prompt | llm | StrOutputParser()
    result = chain.invoke({"input": state["input_data"], "context": state.get("accumulated", "")})
    accumulated = state.get("accumulated", "") + f"\n\n[gather_evidence]: " + result
    print(f"  Step 2/4: gather_evidence")
    return {"accumulated": accumulated, "final_output": result}

def assess_gaps(state: WorkflowState) -> WorkflowState:
    prompt = ChatPromptTemplate.from_messages([
        ("system", "You are step 3 of a LangGraph Compliance Checklist Flow pipeline. "
         "Your role: assess gaps. "
         "Process the input and produce structured output for the next step."),
        ("human", "Input: {input}\nPrevious steps: {context}")
    ])
    chain = prompt | llm | StrOutputParser()
    result = chain.invoke({"input": state["input_data"], "context": state.get("accumulated", "")})
    accumulated = state.get("accumulated", "") + f"\n\n[assess_gaps]: " + result
    print(f"  Step 3/4: assess_gaps")
    return {"accumulated": accumulated, "final_output": result}

def generate_report(state: WorkflowState) -> WorkflowState:
    prompt = ChatPromptTemplate.from_messages([
        ("system", "You are step 4 of a LangGraph Compliance Checklist Flow pipeline. "
         "Your role: generate report. "
         "Process the input and produce structured output for the next step."),
        ("human", "Input: {input}\nPrevious steps: {context}")
    ])
    chain = prompt | llm | StrOutputParser()
    result = chain.invoke({"input": state["input_data"], "context": state.get("accumulated", "")})
    accumulated = state.get("accumulated", "") + f"\n\n[generate_report]: " + result
    print(f"  Step 4/4: generate_report")
    return {"accumulated": accumulated, "final_output": result}
```

## Step 3 — Build and Compile Graph

```python
graph = StateGraph(WorkflowState)
graph.add_node("define_requirements", define_requirements)
graph.add_node("gather_evidence", gather_evidence)
graph.add_node("assess_gaps", assess_gaps)
graph.add_node("generate_report", generate_report)

graph.set_entry_point("define_requirements")
graph.add_edge("define_requirements", "gather_evidence")
graph.add_edge("gather_evidence", "assess_gaps")
graph.add_edge("assess_gaps", "generate_report")
graph.add_edge("generate_report", END)

app = graph.compile()
print("LangGraph Compliance Checklist Flow — workflow compiled!")
```

## Step 4 — Run the Workflow

```python
sample_input = "Analyze and process this request through the LangGraph Compliance Checklist Flow pipeline."
result = app.invoke({
    "input_data": sample_input, "accumulated": "", "final_output": "",
})

print("=== WORKFLOW RESULT ===")
print(result["final_output"])
print("\n=== FULL TRACE ===")
print(result["accumulated"][:1000])
```
